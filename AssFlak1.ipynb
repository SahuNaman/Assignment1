```python
Answer--> Flask is a micro web framework written in Python. 
It is classified as a microframework because it does not require particular tools or libraries.
It has no database abstraction layer, form validation, or any other components where pre-existing third-party libraries provide common functions.
-------
The advantages of Flask Framework are:
----------------------------------------
1.It is a lightweight framework that offers hassle-free development.
2.Provide flexibility to the developer to experiment with their modules or architecture.
3.It is suitable for small projects.
4.Offers a built-in development server and fast debugger.
5.Easily scalable for the applications.
6.Support for secure cookies.
```

```python
Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.
```

```python
Answer ---> The url_for() function is used in Flask for URL Building.

Example ---> 
from flask import *  
   
app = Flask(__name__)  
  
@app.route('/admin')  
def admin():  
    return 'admin'  
  
@app.route('/librarion')  
def librarion():  
    return 'librarion'  
  
@app.route('/student')  
def student():  
    return 'student'  
  
@app.route('/user/<name>')  
def user(name):  
    if name == 'admin':  
        return redirect(url_for('admin'))  
    if name == 'librarion':  
        return redirect(url_for('librarion'))  
    if name == 'student':  
        return redirect(url_for('student'))  
if __name__ =='__main__':  
    app.run(debug = True)  
```
